```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('robot-upflairs.png')
```

```python
def imshow(im):
    plt.imshow(im[:,:,::-1])
    plt.axis('off')
    plt.show()
```

```python
imshow(img)
```

```python
img_res = cv2.resize(img, (200,200), interpolation=cv2.INTER_CUBIC)
print(img_res.shape)
img_res2 = cv2.resize(img, None, fx=2, fy=0.6, interpolation=cv2.INTER_CUBIC)
print(img_res2.shape)
```

```python
from skimage.transform import rotate
```

```python
img_rot = rotate(img, angle=-45, resize=True)
imshow(img_rot)
```

```python
from time import sleep
for a in range(0,361,5):
    img_rot = rotate(img, angle=a, resize=True)
    cv2.imshow('',img_rot)
    cv2.waitKey(1)
    sleep(0.1)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
from time import sleep
import numpy as np
for a in np.arange(0.1,3.0,0.1):
    img_res = cv2.resize(img, None, fx=a, fy=a, interpolation=cv2.INTER_CUBIC)
    cv2.imshow('',img_res)
    cv2.waitKey(1)
    sleep(0.1)
for a in np.arange(3.0,0,-0.1):
    img_res = cv2.resize(img, None, fx=a, fy=a, interpolation=cv2.INTER_CUBIC)
    cv2.imshow('',img_res)
    cv2.waitKey(1)
    sleep(0.1)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
