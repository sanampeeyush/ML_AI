```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('robot-upflairs.png')
img2 = cv2.imread('robot2.png')
img2 = img2[150:150+288, 150:150+288,:]
```

```python
img.shape, img2.shape
```

```python
def showimg(im):
    plt.imshow(im[:,:,::-1])
    plt.axis('off')
    plt.show()

def imshow(im): 
    cv2.imshow('Preview', im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.waitKey(1)  # for macOS only
```

```python
showimg(img)
showimg(img2)
```

```python
added = cv2.add(img, img2)
showimg(added)
```

```python
subtracted = cv2.subtract(img, img2)
showimg(subtracted)
```

```python
import numpy as np
```

```python
img2_1 = cv2.multiply(img2, 2 * np.ones(img2.shape).astype('uint8'))
showimg(img2_1)
```

```python
# Adjusting Contrast(alpha) and brightness(beta)
img2_2 = cv2.convertScaleAbs(img2, alpha = 1.3, beta = 30)
showimg(img2_2)
```

```python
img2_3 = cv2.resize(img2, (400,400), interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
img2_4 = cv2.resize(img2, None, fx=2, fy=0.5, interpolation=cv2.INTER_CUBIC)
img2_4.shape
```

```python
showimg(img2_4)
```

```python
from skimage.transform import rotate
```

```python
img2_5 = rotate(img2, angle=-45, resize=True)
showimg(img2_5)
```

```python
from time import sleep
repeat = True
while repeat:
    for i in range(0,361,5):
        i = rotate(img, angle = i, resize = True)
        cv2.imshow('',i)
        key = cv2.waitKey(1)
        if key == ord('q'):
            repeat = False
            break
        sleep(0.1)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
